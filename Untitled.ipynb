```python
#import dependencies 
import pandas as pd
from bs4 import BeautifulSoup
import requests
import pymongo
from collections import Counter 
from splinter import Browser
executable_path = {"executable_path": "chromedriver.exe"}
browser = Browser("chrome", **executable_path, headless=False)
import time
import matplotlib.pyplot as plt
import sqlalchemy
```

```python
# visit the TDOJ death row page
url = 'https://open.spotify.com/playlist/2YRe7HRKNRvXdJBp9nXFza'
#call the browers and go to the url
browser.visit(url)
```

```python
#grab the value in the browser
spotify_html = browser.html
#convert to a beautiful soup object
spotify_soup = BeautifulSoup(spotify_html, 'html.parser')
```

```python
print(spotify_soup)
```

```python
mydivs = spotify_soup.findAll("div", {"class": "tracklist-name ellipsis-one-line"})
mydivs[99]
```

```python
song_titles = []
for song in mydivs:
    name = song.get_text()
    song_titles.append(name)


```

```python
song_titles
```

```python
#mydivs2 = spotify_soup.findAll("a", {"class": "tracklist-row__artist-name-link"})
#mydivs2

#get second line which includes all artists and album
#mydivs2 = spotify_soup.findAll("div", {"class": "second-line"})
#mydivs2

mydivs2 = spotify_soup.findAll("span", {"class": "TrackListRow__artists ellipsis-one-line"})
mydivs2
```

```python
artist_names = []
for artist in mydivs2:
    name = artist.get_text()
    artist_names.append(name)

```
